```python
import numpy as np
import matplotlib.pyplot as plt


from sklearn.linear_model import LinearRegression
```

## Exercice 1: Utilisation de scikit-learn pour la regression lineaire


```python
#generation de donnees test
n = 100
x = np.arange(n)
y = np.random.randn(n)*30 + 50. * np.log(1 + np.arange(n))

# instanciation de sklearn.linear_model.LinearRegression
lr = LinearRegression()
lr.fit(x[:, np.newaxis], y)  # np.newaxis est utilise car x doit etre une matrice 2d avec 'LinearRegression'

# representation du resultat
fig = plt.figure()
plt.plot(x, y, 'r.')
plt.plot(x, lr.predict(x[:, np.newaxis]), 'b-')
plt.legend(('Data', 'Linear Fit'), loc='lower right')
plt.title('Linear regression')
plt.show()
```

### <span style="color:blue">QUESTION 1.1 :</span> 
Expliquer ce qu'est *lr* et ce que font *lr.fit* et *lr.predict*



### <span style="color:blue">REPONSE 1.1 :</span> 
Ici, tu crées un objet LinearRegression de scikit-learn.
Ce modèle cherche à ajuster une droite de la forme :
y=a⋅x+b
où a est la pente et b l’ordonnée à l’origine.

lr.fit :
le modèle a appris deux choses principales :
lr.coef_ → le(s) coefficient(s) de la droite (la pente, notée a si une seule variable).
lr.intercept_ → l’ordonnée à l’origine (la constante b).
La régression linéaire cherche à approximer :
^y=a⋅x+b

x[:, np.newaxis] → fournit les valeurs de x sous forme de matrice colonne (obligatoire car sklearn attend toujours une matrice 2D).

Pour chaque x_i, le modèle calcule :

𝑦𝑖^=𝑎⋅𝑥𝑖+𝑏

Le résultat est un tableau de la même taille que x, contenant les valeurs estimées de y.
